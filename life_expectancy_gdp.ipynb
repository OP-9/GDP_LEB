<h1>Analysis of Relationship of the Gross Domestic Product (GDP) and the Life Expectancy at Birth of a Nation</h1>
<h2>Points of analysis:</h2>
<ul>
    <li>Country(own) and GDP Growth over time</li>
    <li>Variation over time of life expectancy country vs country</li>
    <li>LEB v GDP</li>
</ul>

```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("all_data.csv")

#print(data.info())
```

```python
#Making it easier to handle the data

data.columns = map(str.lower, data.columns)

data = data.rename(columns={"life expectancy at birth (years)":"leb"})

#data.country.unique()

#data.isnull()
```

<h2>Setting up dataframes by country</h2>

```python
#Data by nation

#def create_country_data(new_df_country, country_name):
    #new_df_country = data[data["country"]==country_name]
    #new_df_country.set_index("country", inplace=True)
    #new_df_country.reset_index(inplace=True)

chile_info = data[data["country"] == "Chile"]
chile_info.set_index("country", inplace=True)
chile_info.reset_index(inplace=True)

china_info = data[data["country"] == "China"]
china_info.set_index("country", inplace=True)
china_info.reset_index(inplace=True)

ger_info = data[data["country"] == "Germany"]
ger_info.set_index("country", inplace=True)
ger_info.reset_index(inplace=True)

mex_info = data[data["country"] == "Mexico"]
mex_info.set_index("country", inplace=True)
mex_info.reset_index(inplace=True)

usa_info = data[data["country"] == "United States of America"]
usa_info.set_index("country", inplace=True)
usa_info.reset_index(inplace=True)

zim_info = data[data["country"] == "Zimbabwe"]
zim_info.set_index("country", inplace=True)
zim_info.reset_index(inplace=True)

```

<h2>Defining Functions</h2>

<h3>Function: Plot GDP against time</h3>

```python
#GDP Over Time

def plot_gdp_per_country(country_year, country_gdp, country_name, color):
    title_info = "GDP over time for " + country_name
    plt.plot(country_year, country_gdp, marker='*', color=color)
    plt.xlabel("Year")
    plt.ylabel("GDP (Trillions of US Dollars)")
    plt.title(title_info)
    plt.show()
    plt.clf()
```

<h3>Function: Plot Life Expectancy at Birth Per Country</h3>

```python
# Function for Life Expectancy Per Country

def plot_leb_per_country(country_year, country_leb, country_name, color):
    title_info = "Life Expectancy at Birth " + country_name + " (years)"
    plt.plot(country_year, country_leb,marker='o', color=color)
    plt.subplot().set_xticks([2000, 2005, 2010, 2015])
    plt.subplot().set_yticks([country_leb[i] for i in range(15, 0, -3)])
    plt.xlabel("Year")
    plt.ylabel("Life Expectancy at Birth")
    plt.title(title_info)
    plt.show()
    plt.clf()
```

<h3>Function: Plot Scatterplot to view relationship between GDP and LEB</h3>

```python
#Function for scatter plot for relationship between gdp and leb

def gdp_and_leb (country_info, country_name):
    sns.lmplot(data=country_info, x="gdp", y="leb")
    plt.plot(color="green")
    plt.xlabel("GDP (Trillion US Dollars)")
    plt.ylabel("Life Expectancy at Birth (in Years)")
    plt.title("Relationship Between Life Expectancy and GDP for "+country_name)
    plt.show()
    plt.clf()
```

<h3>Function: Plot Life Expectancy at Birth</h3>

```python
#Life Expectancy at Birth Between Select Countries

def leb_for_subplot(country_year, country_leb, country_name, color):
    plt.plot(country_year, country_leb, color=color, marker="o")
    plt.title(country_name)
    plt.xlabel("Year")
    plt.ylabel("LEB in Years")

```

<h2>Analysis</h2>

<h3>Determining Average Life Expectancy</h3>

<h3>Comparing Life Expectancy at Birth Between All Countries</h3>

```python
#Comparing Life Expectancy at Birth Between Select Countries
plt.figure(figsize=[14,8])
plt.subplot(3, 2, 1)
leb_for_subplot(chile_info.year, chile_info.leb, "Chile", "Red")

plt.subplot(3, 2, 2)
leb_for_subplot(china_info.year, china_info.leb, "China", "Blue")

plt.subplot(3, 2, 3)
leb_for_subplot(ger_info.year, ger_info.leb, "Germany", "Orange")


plt.subplot(3, 2, 4)
leb_for_subplot(mex_info.year, mex_info.leb, "Mexico", "Green")

plt.subplot(3, 2, 5)
leb_for_subplot(usa_info.year, usa_info.leb, "USA", "Black")

plt.subplot(3, 2, 6)
leb_for_subplot(zim_info.year, zim_info.leb, "Zimbabwe", "gray")


plt.suptitle("Comparing Life Expectancy at Birth Between Select Nations")
plt.subplots_adjust(hspace=0.5, wspace=0.2)

plt.show()
```

<h3>Determining Relationship Between GDP and Life Expectancy for All Countries</h3>

```python
#Comparing Relationship Between GDP and Life Expectancy Growth for All Countries

sns.scatterplot(x=data.gdp, y=data.leb, hue=data.country)
plt.legend()
plt.xlabel("GDP (Trillions of US Dollars)")
plt.ylabel("Life Expectancy at Birth (in Years)")
plt.savefig("gdp_leb_all.png")
plt.show()
plt.clf()

```

```python
#Comparing Above Relationship Amongst All Countries

p=sns.FacetGrid(data=data, col='country', col_wrap=2, hue='country',sharex=False, sharey=False)
p.map_dataframe(sns.scatterplot, x="gdp", y="leb")
p.set_axis_labels("GDP (Trillion USD)", "Life Expectancy at Birth (Years)")
p.add_legend(title="Country")

#p=sns.lmplot(data=data, x="gdp", y="leb",col="country", hue="country",col_wrap=3,facet_kws=dict(sharex=False,sharey=False))
#p.add_legend(title="Countries")
p.set_axis_labels("GDP in Trillion USD", "Life Expectancy at Birth in Years")

plt.savefig("all_countries.png")
```
